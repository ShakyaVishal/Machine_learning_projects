```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=442)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
def model(a):
    ML=a
    ML.fit(x_train,y_train)
    pre=ML.predict(x_test)
    score=cross_val_score(a,x,y,cv=3)
    b=score.mean()
    return(print('Taining_score = ',ML.score(x_train,y_train),'\ncross_val_score = ',b,'\nR2_score = ',r2_score(y_test,pre),'\nMAE = ',mean_absolute_error(y_test,pre),'\nMSE = ',mean_squared_error(y_test,pre),'\nRMSE = ',np.sqrt(mean_squared_error(y_test,pre))))
```

```python
from sklearn.linear_model import LinearRegression
model(LinearRegression())
```

```python
from sklearn.linear_model import Lasso
model(Lasso())
```

```python
from sklearn.linear_model import Ridge
model(Ridge())
```

```python
from sklearn.linear_model import ElasticNet
model(ElasticNet())
```

```python
from sklearn.neighbors import KNeighborsRegressor
model(KNeighborsRegressor())
```

```python
from sklearn.tree import DecisionTreeRegressor
model(DecisionTreeRegressor())
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model(GradientBoostingRegressor())
```

```python
from sklearn.ensemble import RandomForestRegressor
model(RandomForestRegressor())
```

```python
models=[]
models.append(('LinearRegression',LinearRegression()))
models.append(('Lasso',Lasso()))
models.append(('Ridge',Ridge()))
models.append(('ElasticNet',ElasticNet()))
models.append(('DecisionTreeRegressor',DecisionTreeRegressor()))
models.append(('KNeighborsRegressor',KNeighborsRegressor()))
models.append(('GradientBoostingRegressor',GradientBoostingRegressor()))
models.append(('RandomForestRegressor',RandomForestRegressor()))
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
ModelName = []
MSE = []
R2score =[]
cvs=[]
MAE = []
for name, model in models:
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    MSE.append(mean_squared_error(y_test, y_pred))
    R2score.append(r2_score(y_test,y_pred))
    MAE.append(mean_absolute_error(y_test,y_pred))
    kfold = KFold(n_splits=10, random_state=10) 
    re = cross_val_score(model, x, y, cv=kfold, scoring='r2').mean()
    cvs.append(re)
    ModelName.append(name)
result = pd.DataFrame({'Model_name': ModelName, 'Mean_squared_error': MSE ,'R2_score': R2score,'Mean_absolute_error': MAE,'Cross_val_score':cvs})
result
```

I choose Ridge as my final model because it has lowest Mean_squared_error and good R2_score as compared to other model
